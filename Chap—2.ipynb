 ## 1、 we will draw the pic for binomimal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
def Bin_plot(miu,N):
    ind=np.arange(0,N+1)
    C=[]
    for i in ind:   
        C.append(np.divide(np.divide(np.math.factorial(N),np.math.factorial(N-i)),np.math.factorial(i))*miu**i*(1-miu)**(N-i))
    
    plt.plot(ind,C)
    plt.show()
Bin_plot(0.3,100)
```

## 2、here we draw the picture of Bayesian inference shown in 2.1.1

```python
ind=np.arange(0,1,0.01)
def Beta(a,b,ind):
    out=np.math.factorial(a+b-1)/np.math.factorial(a-1)/np.math.factorial(b-1)*np.power(ind,a-1)*np.power(1-ind,b-1)
    return ind,out
ind,prior_out=Beta(3,3,ind)

plt.plot(ind,prior_out)
plt.show()

ind,liklihood_out=Beta(13,2,ind)
plt.plot(ind,liklihood_out)##how to get the
plt.show()

ind,post_out=Beta(16,5,ind)
plt.plot(ind, post_out)
plt.show()
```

## 4、Maximize likelihood function demo
- here we carried out $\log p(X|\theta)=\sum_{i=1}^{i=N}\log p(x_i|\theta)$, and found that it reaches maximum around mean and std.

```python
mean=10
std=3
data=np.random.normal(loc=mean,scale=std,size=(1000)) ##cale is the standard deviation
x=np.arange(5,15,0.1)
y=np.arange(1,10,0.1)
xx,yy=np.meshgrid(x,y)

data_epDim= np.repeat(data[:, np.newaxis], xx.shape[0], axis=1)
data_epDim = np.repeat(data_epDim[:,:, np.newaxis], xx.shape[1], axis=2)
likelihood=np.log(1/np.sqrt(2*3.1415*yy**2)*np.exp(-(data_epDim-xx)**2/2/yy**2))
print(likelihood.shape)
nn=np.sum(likelihood,axis=0)
fig = plt.figure()
ax = fig.gca(projection='3d')

print(nn.shape)
ax.plot_surface(xx,yy,nn)
plt.show()
```

### as you can see that, likelihood function reach its maximum around (10,3)
- it means that likelihood function is a distribution, and it can be integrated.And sometimes, we use maximum posterior(MAP) to determine the most probable values of paramter instead of using the distribution itself.

- Thus due to the bayesian theorem, we can get the posterior distribution $ p(\theta | X)=\frac{p(X|\theta)p(\theta)}{\int p(X|\theta)d\theta}$,but how to integrate? here we know $p(X|\mu,\sigma)$ is a function of $\theta,\sigma $,actually, we can integrate it.

## 6、 draw the contour depicted in Fig. 2.8


```python
r=1
%matplotlib notebook

def plot_contour(r,miu,sig):
    w, v = np.linalg.eig(sig)
    xx_curv1=np.arange(-np.sqrt(w[0])*np.sqrt(r),np.sqrt(w[0])*np.sqrt(r),0.01)
    xx_curv2=np.flip(xx_curv1,0)
    #print(xx_curv2)
    xx_curv=np.append(xx_curv1,xx_curv2)
    yy_curv1=np.sqrt((r-np.square(xx_curv1)/w[0])*w[1])
    yy_curv2=-np.sqrt((r-np.square(xx_curv2)/w[0])*w[1])
    yy_curv=np.append(yy_curv1,yy_curv2)
    zz=np.stack((xx_curv,yy_curv),1)
    zz_orig=(v@zz.T).T+miu
    plt.plot(zz_orig[:,0],zz_orig[:,1])
    #print(zz_orig.shape)
    
    zz=np.sum((zz_orig-miu)@sig_*(zz_orig-miu),1)
    #print(zz.shape)
    N_curv=1/2/3.1415**2*1/np.linalg.det(sig)**0.5*np.exp(-0.5*zz)
    zz=np.stack((zz_orig[:,0],zz_orig[:,1],N_curv),1)
    print(zz.shape)
    return zz
    
fig = plt.figure()


for i in range(1,90):
    ff=plot_contour(i*0.1,miu,sig)
    ax = fig.gca(projection='3d')
    #ax.plot_surface(xx,yy,N)
    ax.plot(ff[:,0],ff[:,1],ff[:,2])
    #ax.plot(ff[:,0],ff[:,1])
plt.show()
```

###  here if we consider the dimensionality, for i.i.d. ,with the increase of dimension,  probablity density is identical on sphere of radius r in N dimension, and with N increasing, probablity density peaks increase.
